# Autoresearch Classifier — Experiment Analysis

Visualize experiment results from `results.tsv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results.tsv", sep="\t")
print(f"Total experiments: {len(df)}")
print(f"  keep:    {(df.status == 'keep').sum()}")
print(f"  discard: {(df.status == 'discard').sum()}")
print(f"  crash:   {(df.status == 'crash').sum()}")
print()
print(f"Best val_accuracy: {df.val_accuracy.max():.6f}")
best_row = df.loc[df.val_accuracy.idxmax()]
print(f"  commit:      {best_row.commit}")
print(f"  val_f1:      {best_row.val_f1:.6f}")
print(f"  description: {best_row.description}")
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

colors = {"keep": "#2ecc71", "discard": "#e74c3c", "crash": "#95a5a6"}

for status, group in df.groupby("status"):
    ax.scatter(group.index, group.val_accuracy,
              c=colors.get(status, "gray"), label=status, s=60, zorder=3)

kept = df[df.status == "keep"]
if len(kept) > 0:
    ax.plot(kept.index, kept.val_accuracy, c="#2ecc71", alpha=0.5, linewidth=1.5, zorder=2)
    running_best = kept.val_accuracy.cummax()
    ax.step(kept.index, running_best, c="#27ae60", linewidth=2, linestyle="--",
            label="best so far", where="post", zorder=2)

ax.set_xlabel("Experiment #")
ax.set_ylabel("val_accuracy")
ax.set_title("Autoresearch Classifier — Experiment Progress")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("progress.png", dpi=150)
plt.show()
print("Saved progress.png")
```

```python
print("=== All experiments ===")
print(df.to_string(index=True))
print()

baseline_acc = df.val_accuracy.iloc[0] if len(df) > 0 else 0
best_acc = df.val_accuracy.max() if len(df) > 0 else 0
improvement = best_acc - baseline_acc
print(f"Baseline accuracy: {baseline_acc:.6f}")
print(f"Best accuracy:     {best_acc:.6f}")
print(f"Improvement:       {improvement:+.6f}")
```
